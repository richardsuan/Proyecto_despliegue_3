```python
import pandas as pd

# Ruta del archivo Excel
file_path = './content/tabla_df.xlsx'

# Cargar el archivo Excel, reemplazando "N.A." por NaN
df = pd.read_excel(file_path, na_values="N.A.")

# Mostrar las primeras filas del archivo cargado
df.head()
```

```python
# Obtener las frecuencias de la variable categórica 'rango de precios'
frecuencia_rangos = df['rango_precios_var_dependiente'].value_counts()

# Mostrar la frecuencia de cada categoría en 'rango de precios'
print(frecuencia_rangos)
```

```python
# Obtener la proporción de cada categoría en 'rango de precios'
proporcion_rangos = df['rango_precios_var_dependiente'].value_counts(normalize=True) * 100

# Mostrar las proporciones (en porcentaje) de cada categoría
print(proporcion_rangos)
```

```python
# Frecuencia de 'rango de precios' agrupada por departamento
frecuencia_por_departamento = df.groupby('departamen')['rango_precios_var_dependiente'].value_counts()

# Mostrar la frecuencia por departamento
print(frecuencia_por_departamento)

# Frecuencia de 'rango de precios' agrupada por municipio
frecuencia_por_municipio = df.groupby('municipio')['rango_precios_var_dependiente'].value_counts()

# Mostrar la frecuencia por municipio
print(frecuencia_por_municipio)
```

```python
# Resumen estadístico de variables numéricas
resumen_numericas = df.describe()
resumen_numericas
```

```python
# Encuentra las columnas categóricas de tipo 'object' o 'category'
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Muestra las columnas categóricas y su cantidad
print("Columnas categóricas:", categorical_columns)
print("Número de columnas categóricas:", len(categorical_columns))

# Elimina las columnas categóricas del DataFrame
df_numeric = df.drop(columns=categorical_columns)

# Muestra el DataFrame sin las columnas categóricas
print(df_numeric.head())
```

```python
# Calcula la matriz de correlación
correlation_matrix = df_numeric.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcula la matriz de correlación

# Configura el tamaño del gráfico
plt.figure(figsize=(20, 15))

# Dibuja la heatmap de la matriz de correlación
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Muestra el gráfico
plt.title('Matriz de Correlación')
plt.savefig('matriz_correlacion.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df_numeric.hist(figsize=(20, 20), bins=50)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Obtener las frecuencias de la variable categórica 'departamen'
frecuencia_departamentos = df['departamen'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
sns.barplot(x=frecuencia_departamentos.index, y=frecuencia_departamentos.values)

# Configurar etiquetas y título
plt.title('Frecuencia de Departamentos en el Dataset')
plt.xlabel('Departamento')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.savefig('departamentos.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sns.boxplot(data=df_numeric)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.decomposition import PCA
import numpy as np
pca = PCA(n_components=39)
pca.fit(df_numeric)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Número de Componentes')
plt.ylabel('Varianza Explicada Acumulada')
plt.show()
```
